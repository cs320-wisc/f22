## Paths

A **path** is a sequence of edges to get from one node to another

````python
g = Graph()
g.edge("A", "B")
g.edge("B", "C")
g.edge("C", "D")
g.edge("D", "F")
g.edge("A", "F")
g.node("E")
g
# no path between A and E
# B => A => F is a path
# B => C => D => F is a path
# future: what is the SHORTEST path between two nodes
# "connected" means there is a path between any two nodes
# "cycle" means a path with same start+end, no repeating edges
# for example: A,B,C,D,F,A
````

## Directed Graphs, DAGs

````python
g = Digraph()
g.edge("A", "B")
g.edge("B", "C")
g.edge("C", "D")
g.edge("D", "F")
g.edge("A", "F")
g.edge("A", "E")
g.node("E")
g
# "parent" means the outgoing end of an edge
# "child" means the incoming end of an edge
# "strongly connected" (same as "connected") => there's a path from any start to any end
# "weakly connected" => it would be connected if we ignored the direction of edges
````

## Trees

**Tree**: DAG where one node (the **root**) has no parents and all others have exactly one parent

````python
# "root" is any node with no parents
g = Digraph()
g.edge("A", "B")
g.edge("B", "C")
g.edge("C", "D")
#g.edge("D", "F")
g.edge("A", "F")
g.edge("A", "E")
g.node("E")
g
# "leaf" is any node with no children
````
